## Text Preprocessing

```python
# Text Lowercase:
df1['clean_overview'] = df1['Overview'].str.lower()
df1.head()
```

```python
# Removing numbers and punctuations
df1["clean_overview1"] = df1['clean_overview'].apply(lambda x: re.sub('[^a-zA-Z]',' ',x))
df1.head()
```

```python
# Removing unwanted space:

df1["clean_overview2"] = df1["clean_overview1"].apply(lambda x: re.sub('\s+', ' ',x))
df1.head()
```

```python
# Tokanization
```

```python
df1['Tokenize'] = df1['clean_overview2'].apply(lambda x: nltk.word_tokenize(x))
df1.head()
```

```python
stop_words = nltk.corpus.stopwords.words('english')
Word_List = []
for sentence in df1["Tokenize"]:
    temp = []
    for word in sentence:
        if word not in stop_words and len(word) >= 3:
            temp.append(word)
    Word_List.append(temp)
```

```python
df1['Clean_Tokenize'] = Word_List 
```

```python
df1['Clean_Tokenize']
```

```python
# Combining the words in Genre/Actors and Director columns into single words:
df1['Genre'] = df1['Genre'].apply(lambda x: x.split(','))
df1["actors"] = df1["actors"].apply(lambda x: x.split(',')[:4])
df1["Director"] = df1["Director"].apply(lambda x: x.split(','))
```

```python
def clean_text(sentence):
    temp = []
    for word in sentence:
        temp.append(word.lower().replace(' ', ''))
    return temp
```

```python
df1["Genre"] = [clean_text(x) for x in df1['Genre']]
df1["actors"] = [clean_text(x) for x in df1['actors']]
df1["Director"] = [clean_text(x) for x in df1['Director']]
```

```python
df1.head()
```

```python
# Combining all columns into single sentence:

columns = ['Clean_Tokenize', 'Genre', 'actors', 'Director']
sentence_list = []
for i in range(len(df1)):
    words = ''
    for col in columns:
        words += ' '.join(df1[col][i]) + ' '
    sentence_list.append(words)
sentence_list[0]
```

```python
df1['Clean_input'] = sentence_list
```

```python
df1.head()
```

```python
df2 = df1[['Series_Title', 'Clean_input']]
df2.head(20)
```

## Feature Extraction

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
tfidf = TfidfVectorizer()
features = tfidf.fit_transform(df2['Clean_input'])
features
```

```python
features.toarray().shape
```

## Cosine Similarity:

```python
# Cosine similarity matrix
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim = cosine_similarity(features, features)
print(cosine_sim)
```

```python
cosine_sim.shape
```

## Recommendation System

```python
index = pd.Series(df2['Series_Title'])
index.head()
```

```python
pd.Series(cosine_sim[0]).sort_values(ascending = False).head(6)
```

```python
list_of_index = pd.Series(cosine_sim[0]).sort_values(ascending = False).head(6).index
list_of_index
```

```python
for idx in list_of_index[1:]:
    print(index[idx])
```

```python
movie_idx = index[index == "The Dark Knight"].index[0]
score = pd.Series(cosine_sim[movie_idx]).sort_values(ascending = False)
top5_movie_idx = list(score.iloc[1:6].index)

list_of_recommended_movies = [index[idx] for idx in top5_movie_idx]
list_of_recommended_movies
```

```python
# Resuable function

def recommendation_system(movie_title):
    movie_idx = index[index == movie_title].index[0]
    score = pd.Series(cosine_sim[movie_idx]).sort_values(ascending=False)
    top5_movie_idx = list(score.iloc[1:6].index)
    list_of_recommended_movies = [index[idx] for idx in top5_movie_idx]
    return list_of_recommended_movies
```

```python
recommendation_system('The Lord of the Rings: The Fellowship of the Ring')
```
